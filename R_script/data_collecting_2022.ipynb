```r
doc <- read_html("https://movieweb.com/movies/2022")
nodes <- xml_find_all(doc, "//div[@class='database-card-title']//a")
link_list <- xml_attr(nodes, "href")
```

```r
make_link <- function(link){
    simplified_link <- sub("^(.{1,2}).(.*)", "\\1\\2", link)
    simplified_link <- sub("^(.{1,2}).(.*)", "\\1\\2", simplified_link)
    simplified_link <- sub("^(.{1,2}).(.*)", "\\1\\2", simplified_link)
    simplified_link <- sub("^(.{1,2}).(.*)", "\\1\\2", simplified_link)
    simplified_link <- gsub("-", "_", simplified_link)
    adresses <- paste0("https://www.rottentomatoes.com/",simplified_link)
    return(adresses)
}
```

```r
length_list <- nrow(output)
movie_name <- c()
rating <- c()
year <- c()
types <- c()
duration <- c()
audience_rating <- c()
expert_rating <- c()
for(i in c(1:length_list)){
  adresse <- make_link(link_list[i])
  result <- tryCatch(find_rating_movie_direct(adresse), error = function(e) NULL)
  if (is.null(result) || length(result) == 0) {
    next
  }
  print(result$`Movie Name`)
  movie_name <- append(movie_name, coalesce(result$`Movie Name`, ""))
  rating <- append(rating, coalesce(result$`Rating`, ""))
  year <- append(year, coalesce(result$`Year`, ""))
  types <- append(types, coalesce(result$`Types`, ""))
  duration <- append(duration, coalesce(result$`Duration`, ""))                  
  audience_rating <- append(audience_rating, coalesce(result$`Audience Score`, ""))
  expert_rating <- append( expert_rating ,coalesce(result$`Tomatometer Score`, ""))
}
```

```r
output_2022 <- data.frame(movie_name,rating,year,types,duration,audience_rating,expert_rating)
```

```r
write.csv(output_2022,"rotten_tomatoes_movies_2022.csv", row.names = TRUE)
```
