```r
data_movie_raw <- read.csv(file = 'rotten_tomatoes_movies.csv', header = TRUE)
str(data_movie_raw)
```

```r
dropped_columns <- c( # les colonnes à retirer
    'movie_info',
    'critics_consensus',
    'directors',
    'authors',
    'actors',
    'production_company',
    'tomatometer_count',
    'audience_count',
    'tomatometer_top_critics_count',
    'tomatometer_fresh_critics_count',
    'tomatometer_rotten_critics_count'
)
```

```r
data_movie = data_movie_raw[,!(names(data_movie_raw) %in% dropped_columns)]
data_movie
write.csv(data_movie,"data_movie_cleaned.csv", row.names = TRUE)
```

```r
dropped_columns_2 <- c(
    'audience_status',
    'tomatometer_status',
    'streaming_release_date',
    'rotten_tomatoes_link',
    'movie_info',
    'critics_consensus',
    'directors',
    'authors',
    'actors',
    'production_company',
    'tomatometer_count',
    'audience_count',
    'tomatometer_top_critics_count',
    'tomatometer_fresh_critics_count',
    'tomatometer_rotten_critics_count'
)
```

```r
transform_duration <- function(duration) {
  duration <- as.numeric(duration)  # convert duration to numeric if needed
  # Check if the duration is NA, an empty string, or "NA"
  if (is.na(duration) || duration == "" || duration == "NA") {
    return("")
  }
  
  hours <- floor(duration / 60)
  minutes <- duration - hours * 60
  return(paste(hours, "h", minutes, "m", sep=""))
}
```

```r
data_movie = data_movie_raw[,!(names(data_movie_raw) %in% dropped_columns_2)]

# First, let's rename the columns
colnames(data_movie) <- c("movie_name", "rating", "types", "year", "duration", "audience_rating", "expert_rating")

# Now let's transform the movie_name column
data_movie$movie_name <- toupper(data_movie$movie_name)

# Next, let's extract just the year from the year column
data_movie$year <- as.integer(substr(data_movie$year, 1, 4))

# Now let's transform the duration column
data_movie$duration <- lapply(data_movie$duration, transform_duration)

# Now let's reorder the columns to match the final result column order you specified
data_movie <- data_movie[c("movie_name", "rating", "types", "year", "duration", "audience_rating", "expert_rating")]
```
